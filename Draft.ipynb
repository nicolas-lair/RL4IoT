```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
start_eps=0.95
min_eps=0.01
eps_decay=4000
    
def eps(steps):
    return min_eps + (start_eps - min_eps) * np.exp(-1. * steps / eps_decay)    
```

```python
plt.plot(range(20000), [eps(s) for s in range(20000)])
```
